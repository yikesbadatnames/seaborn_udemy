```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
# Assigning Data

print(f"""
Here are the names of all the data sets in 
seaborn

{sns.get_dataset_names()}


""")
tips = sns.load_dataset("tips")

tips.head()
```

```python
#matplot lib example
plt.bar(x=tips["smoker"], height=tips["tip"], data =tips)
plt.xlabel("smoker")
plt.ylabel("total_bill")
plt.title("Tips/Smoker")
plt.show()
```

```python
# Seaborn Example
sns.barplot(x="smoker", y="tip", data = tips)
plt.title("Tips vs Smoker")
plt.show()
```

```python
# Object Oriented Example
fig1, axes1 = plt.subplots()
sns.barplot(x="smoker", y="total_bill", data = tips, hue="sex",
    estimator= np.sum, errcolor="k", errwidth=0.75, capsize=0.2, ax = axes1)
axes1.set_xlabel("Smoker - Yes/No")
axes1.set_ylabel("Bill Amount")
axes1.set_title("Total Bill v. Smoker")

```

```python
# So, how do we customize our defaults you may so humbly ask...
sns.scatterplot("total_bill", "tip", data= tips)
plt.title("Default Theme - Darkgrid")
plt.show()
```

```python
# By using sns.set_styles, you can go into the back end and mess around with the default settings.

sns.set_style(
    style="ticks", rc = {"axes.edgecolor": "b", "text.color":"r",
    "xtick.color" : "r", "ytick.color":"r"
})

sns.scatterplot('total_bill',"tip", data=tips)
plt.title("New theme")
plt.show()
```
