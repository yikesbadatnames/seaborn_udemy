```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
# Assigning Data

print(f"""
Here are the names of all the data sets in 
seaborn

{sns.get_dataset_names()}


""")
tips = sns.load_dataset("tips")

tips.head()
```

```python
#matplot lib example
plt.bar(x=tips["smoker"], height=tips["tip"], data =tips)
plt.xlabel("smoker")
plt.ylabel("total_bill")
plt.title("Tips/Smoker")
plt.show()
```

```python
# Seaborn Example
sns.barplot(x="smoker", y="tip", data = tips)
plt.title("Tips vs Smoker")
plt.show()
```

```python
# Object Oriented Example
fig1, axes1 = plt.subplots()
sns.barplot(x="smoker", y="total_bill", data = tips, hue="sex",
    estimator= np.sum, errcolor="k", errwidth=0.75, capsize=0.2, ax = axes1)
axes1.set_xlabel("Smoker - Yes/No")
axes1.set_ylabel("Bill Amount")
axes1.set_title("Total Bill v. Smoker")

```

# Plot Styling 
_____________________

```python
# rcs of curent seaborn style
print(f"""
Here are the sns.axes_style() default settings for asethtics. 

{sns.axes_style()}
"""
)

```

```python
# So, how do we customize our default aesthetics you may so humbly ask...
sns.scatterplot("total_bill", "tip", data= tips)
plt.title("Default Theme - Darkgrid")
plt.show()
```

```python
# 
# By using sns.set_styles, you can go into the back end and mess around with the default aesthetics  settings.

sns.set_style(
    style="ticks", rc = {"axes.edgecolor": "b", "text.color":"k",
    "xtick.color" : "g", "ytick.color":"g"
})

sns.scatterplot('total_bill',"tip", data=tips)
plt.title("New theme")
plt.show()
```

```python
# how to temporarily change Styling
with sns.axes_style(style="whitegrid", rc={"font.family": "serif", "font.serif":"Times"}):
    sns.scatterplot("total_bill", "tip", data= tips)
plt.title("White Grid Theme")
plt.show()
```

# Scaling Plot Elements
_________________________

```python
# rcs of current seaborn context
print(f"""
Here are the sns.plotting_context() default settings for asethtics. 

{sns.plotting_context()}
"""
)

```

```python
# example of 'poster' context
sns.set_context(context="poster")
sns.scatterplot("total_bill","tip", data = tips)
plt.title("Poster Context")
plt.show
```

```python
# example of doing one with out the context
with sns.plotting_context(context="poster", rc={"axes.labelsize":20.0,"axes.titlesize":20.0, "grid.linewidth":2.5}):
    sns.scatterplot("total_bill","tip", data=tips)

plt.title("Poster Context")
plt.show()
```

# Can we do both in one freakiiiiing function????
# YOU BET!
---------------------------------

```python
# the set function is how you change both at the same time, but you need 
# axes_style() and plotting_context() to know what to change. 
sns.set(style="white", context= "paper", rc={"axes.edgecolor": "b", "axes.titlesize":20.0})
sns.scatterplot("total_bill","tip", data = tips) 
plt.show()
```

```python
# how do i get rid of these super lame borders?
sns.set(style="ticks")
sns.scatterplot("total_bill","tip", data= tips)
sns.despine(left= True, bottom= True) # despine only normally takes off the top
plt.title("yadaha yeadahaaha")
plt.show()
```

# Great, but now i've made all these crazy changes and i need to make everything normal again.

```python
# All you must do is...
sns.set()

# then call your plot
sns.scatterplot("total_bill","tip", data=tips)
plt.show()
```
