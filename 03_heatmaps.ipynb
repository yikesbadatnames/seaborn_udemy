# What is a heat map?
A heat map is a graphical representation of data which represents data values using colors. 
<br>
Heat maps make it easy to visualize complex data and are useful for idenitfying patterns and areas of concentrations.
<br>
They are used to show the relation between two variables, plotted on x and y axis. The variables plotted on each axis
<br>
can be of any type, categorical lables or numerical values. When plotting categorical variables on the axis, it is good to properly order the 
<br>
variables based on some criteria in order to reveal the patterns in the data.

<br></br>
We can use __categorical color__ to represent __categorical data__, while __numberical data__ requires a __divergent color scale__

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# creating random data
x = np.random.rand(5,5)
df = pd.DataFrame(x)
df
```

```python
# creating heat map
sns.heatmap(df)
plt.show()
```

```python
# first time example
fight_df = sns.load_dataset("flights")
fight_df.head()
```

```python
# pivoting table for heat map
flights = pd.pivot_table(data=fight_df, index = "month", columns = "year", values="passengers")
flights
```

```python
# creating heatmap
my_palette = sns.color_palette("Blues_r")
sns.set_palette(palette= my_palette, n_colors= 15)
sns.heatmap(data=flights, annot = True, fmt="d", cbar = True)
plt.title("flights data")
plt.xlabel("year")
plt.ylabel("month")
plt.show()
```

```python
# assigning a center to the heatmap 
# creating heatmap
my_palette = sns.color_palette("Blues_r")
sns.set_palette(palette= my_palette, n_colors= 15)
sns.heatmap(data=flights, center = flights.loc["Jul",1960])
plt.title("flights data")
plt.xlabel("year")
plt.ylabel("month")
plt.show()
```

```python
# making a square heat map
sns.heatmap(data=df, square=True, xticklabels=2, yticklabels=2,cbar_kws={"orientation":"horizontal"})
plt.show()
```

basically only grabbing one half of the heat map. 

```python
# loading data set.
titanic = sns.load_dataset('titanic')
titanic["survived"].dropna().copy()
titanic
```

```python
titan_heatmap = pd.pivot_table(data = titanic, index="sex", columns = "class", values="survived", aggfunc= np.sum).fillna(0)
titan_heatmap
```

```python
fig, ax = plt.subplots()

sns.heatmap(data = titan_heatmap, cmap= sns.color_palette("Reds_r"),
    ax =ax, linewidths=2, linecolor = "black")
```
